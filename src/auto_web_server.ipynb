````python
import time
import os
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.chrome.options import Options

# === CONFIG ===
CSV_FILE_PATH = "/Users/alichemkhi/Desktop/myProjects/2D_gast/output/v0.2/2D_gast_grouped_v0.2/2D_gast_grouped_v0.2_BMP_Epi.462.csv"  # Replace this
WAIT_TIME = 90  # Time to wait for results to load (adjust based on dataset size)

# === SETUP SELENIUM DRIVER ===
options = Options()
options.headless = False  # Set to True if you want to run in the background
prefs = {"download.prompt_for_download": False}
options.add_experimental_option("prefs", prefs)

driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=options)

for path,path_st,name in zip(file_paths,st_txt_paths,basenames):
    print(name,":",path)
    try:
        # STEP 1: Open MIIC

        driver.get("https://miic.curie.fr/workbench.php")
        time.sleep(5)


        upload_input = driver.find_element(By.XPATH, "/html/body/div[2]/form/div[1]/div/div[2]/div/input")
        upload_input.send_keys(name)
        time.sleep(2)

        
        upload_input = driver.find_element(By.XPATH, "/html/body/div[2]/form/div[1]/div/div[3]/div/input")
        upload_input.send_keys(email)
        time.sleep(2)


        upload_input = driver.find_element(By.XPATH, "/html/body/div[2]/form/div[1]/div/div[4]/div[1]/input[2]")
        upload_input.send_keys(path)
        time.sleep(2)


        upload_input = driver.find_element(By.XPATH, "/html/body/div[2]/form/div[2]/div[4]/div/div[3]/div[1]/input[1]")
        upload_input.send_keys(path_st)
        time.sleep(2)

        
        # STEP 4: Launch MIIC
        print("Launching MIIC job...")
        submit_button = driver.find_element(By.XPATH, "/html/body/div[2]/form/div[2]/div[8]/div/div/input")
        submit_button.click()
        time.sleep(10)


        result_link_element = driver.find_element(By.XPATH, "/html/body/div[2]/div/div[2]/div[3]/div[2]/div/table/tbody/tr[1]/td[2]/a")
        result_link = result_link_element.get_attribute("href")
        print(f"----> Result link: {result_link}")



    except : 
        print("Done.")
        #driver.quit()

````
