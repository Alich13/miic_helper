```python
genes="OSTMI, PDSS2, SOBP, ZBTB24, SMP D2, C6ORF203, ORSLI, LINCO0222, A7GS, SNX3, OSTM1-AS1, BEND3, PPIL6, LOC105377924, SCML4, CBP3121, ARMC2, ARMEZ-ASL, MIRSS, KINAPL, AFGL, SESNI, CRYBG1, CCDC162P, CD24, FOXO3, CD164, SEC63, LINC02532, LINCO2526, PRDM1, MICALL, NR2E1, AK9, FIG4"
geneList=genes.split(",")
geneListStrip=[g.strip() for g in geneList]
geneListStrip
with open("/Users/alichemkhi/Desktop/myProjects/surgery-GBM/output/amena.tsv","w") as file :
    for g in geneListStrip : 
        file.write(f"{g}\n")
```

```python
random.seed(42)  # Fix the seed
# Open and read the JSON file
with open('/Users/alichemkhi/Downloads/saved_graph_7.json', 'r') as file:
    data = json.load(file)

mosaic_df=pd.read_csv("/Users/alichemkhi/Desktop/myProjects/surgery-GBM/output/MI_tables_Tumor-myeloid/causalCCC_df.87-90_V0.csv")
st_df=pd.read_csv("/Users/alichemkhi/Desktop/myProjects/surgery-GBM/output/MI_tables_Tumor-myeloid/causalCCC_st.87-90_V0.csv",sep="\t")

ligands=set(st_df[st_df["group"]=="ligands"]["var_names"])
receptors=set(st_df[st_df["group"]=="receptors"]["var_names"])

linkers = receptors.union(ligands)


first_row=mosaic_df.iloc(axis=0)[0]
last_row=mosaic_df.iloc(axis=0)[-1]
left_names=list(first_row[first_row.notna()].index)
right_names=list(last_row[last_row.notna()].index)

#X2610035D17Rik

print(data["savedPositions"])
new_coordinates = data["savedPositions"]


# random squares where to put the regular genes  
left_square=[[0,900],[0,1500]]
right_square=[[1300,3000],[0,1500]]


gap=50
left_x= 1000 # const
right_x= 1200 # const
left_y= 1500
right_y= 1500

# left cell
for k in left_names  : 
    
    if k in linkers : 
        left_y=left_y-gap
        new_coordinates[k][0]=left_x
        new_coordinates[k][1]=left_y
    else : 
        new_coordinates[k][0] = random.randint(left_square[0][0], left_square[0][1]) 
        new_coordinates[k][1] = random.randint(left_square[1][0], left_square[1][1]) 


# right cell        
for k in right_names  : 
    if k in linkers : 
        right_y=right_y-gap
        new_coordinates[k][0]=right_x
        new_coordinates[k][1]=right_y
    else : 
        new_coordinates[k][0] = random.randint(right_square[0][0], right_square[0][1]) 
        new_coordinates[k][1] = random.randint(right_square[1][0], right_square[1][1]) 

data["savedPositions"]=new_coordinates



with open('/Users/alichemkhi/Downloads/saved_graph_7_modified.json', 'w') as file:
    json.dump(data, file, indent=4)
```
